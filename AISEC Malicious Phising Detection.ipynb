```python
import re

def check_ip(url):
    # Regular expression patterns to match IPv4 and IPv6 addresses
    ipv4_pattern = r'\b(?:\d{1,3}\.){3}\d{1,3}\b'
    ipv6_pattern = r'^(?:(?:[A-F0-9]{1,4}:){7}[A-F0-9]{1,4}|' \
                   r'(?=(?:[A-F0-9]{0,4}:){0,7}[A-F0-9]{0,4}$)' \
                   r'(([0-9A-F]{1,4}:){1,7}|:)(:[0-9A-F]{1,4}){1,7})$'

    if re.search(ipv4_pattern, url) or re.search(ipv6_pattern, url):
        return 1
    else:
        return 0

data['is_ip'] = data['url'].apply(check_ip)
```

```python
data
```

```python
def count_dot(url):
  count = url.count('.')
  return count

data['dot_number'] = data['url'].apply(count_dot)
```

```python
def count_slash(url):
  count = url.count('/')
  return count

data['slash_number'] = data['url'].apply(count_slash)
```

```python
def suspicious_words(url):
  match = re.search('PayPal|login|signin|bank|transfer|account|connect|free|lucky|service|bonus|ebay ',url)
  if match :
    return 1
  else:
    return 0
```

```python
data['sus'] = data['url'].apply(lambda word : suspicious_words(word))
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['type'] = le.fit_transform(data['type'])
```

```python
data
```

```python
data['type'].value_counts()
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
X = data.drop(columns = ['url','type'], axis = 1)
Y = data[['type']]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
model = xgb.XGBClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))
```

```python
from sklearn.model_selection import cross_val_score
cross_val = cross_val_score(model, X, Y, cv = 5)
```

```python
import numpy as np
print("Mean of cross-validation:", np.mean(cross_val))
```
